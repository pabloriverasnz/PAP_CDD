# Eliminación de palabras innecesarias

```python
def clean_text(data, row, column):
    
    symbols = ['%', '(', ')', '#', ':', ',', 'id', '.', ';', 'las', 'los', 'de', 'la', 'lo', 'el', 'la', '-']
    
    text = data.iloc[row,column]
    
    tokenized_text = nltk.word_tokenize(text)
    
    text_nosym = [word for word in tokenized_text if word not in symbols]
    
    cleaned_text_p = [x for x in text_nosym if not any(c.isdigit() for c in x)]
    
    clenaed_text = " ".join(cleaned_text_p)
    
    return clenaed_text
```

```python
cleaned_text = [clean_text(data2, i, 3) for i in range(len(data))]
cleaned_text
```

# Frecuencia de palabras

```python
desc_list = data.iloc[:,3]

word_list_nc = [nltk.word_tokenize(t) for t in desc_list.to_list()]
word_list2 = [val for sublist in word_list_nc for val in sublist]

word_list2
```

```python
word_list2c = [word for word in word_list2 if word not in symbols]
word_list2c
```

```python
word_list2c = [x for x in word_list2c if not any(c.isdigit() for c in x)]
word_list2c
```

```python
# Frequencia de todas las palabras de la columna
frequency = pd.Series(word_list2c).value_counts().sort_values(ascending=False).to_dict()

# Lista de palabras
freq_words = list(frequency.keys())

# Palabras eliminando stop words
filtered_words = [word for word in freq_words if word not in stopwords.words('spanish')]

# Frecuencia de las palabras eliminando stop words
filtered_word_freq = [frequency[k] for k in filtered_words if k in frequency]
```

```python
# DataFrame de la frecuencia de las palabras eliminando stop words
df = pd.DataFrame(data = {'Frecuencia': filtered_word_freq}, index = filtered_words)
df
```
